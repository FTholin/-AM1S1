
##  Calculateur de facture d’électricité

Vous devez développer un programme en langage C pour calculer la facture d’électricité d’un utilisateur, selon sa consommation en kilowattheures (kWh). Le tarif appliqué dépend de la tranche de consommation comme suit :

- **Jusqu’à 100 kWh** : 0,15 € par kWh
- **De 101 à 500 kWh** : 0,20 € par kWh
- **Au-delà de 500 kWh** : 0,25 € par kWh

Le programme devra :
1. Demander à l’utilisateur de saisir sa consommation électrique en kWh.
2. Vérifier que la consommation saisie est valide (elle ne peut pas être négative).
3. Calculer le montant total de la facture en appliquant les tarifs ci-dessus.
4. Afficher le montant total de la facture avec deux décimales.


### **Détails techniques**
1. **Créer une fonction `calculer_facture`** :
   - Cette fonction doit :
     - Prendre en entrée un nombre réel représentant la consommation en kWh.
     - Retourner un nombre réel correspondant au coût total de la facture.
   - Appliquer les tarifs en utilisant des conditions pour déterminer la tranche de consommation.

2. **Gérer les erreurs d’entrée** :
   - Si l’utilisateur entre une valeur négative, afficher un message d’erreur approprié et terminer le programme.

3. **Afficher la facture** :
   - Afficher le montant total en euros, formaté avec deux décimales.

---

### **Exemple d’exécution**

#### Cas 1 : Consommation de 50 kWh
```
Veuillez entrer votre consommation électrique en kWh : 50
Votre facture d'électricité est de : 7.50 €
```

#### Cas 2 : Consommation de 150 kWh
```
Veuillez entrer votre consommation électrique en kWh : 150
Votre facture d'électricité est de : 25.00 €
```

#### Cas 3 : Consommation de 600 kWh
```
Veuillez entrer votre consommation électrique en kWh : 600
Votre facture d'électricité est de : 130.00 €
```

#### Cas 4 : Consommation négative
```
Veuillez entrer votre consommation électrique en kWh : -20
Erreur : la consommation ne peut pas être négative.
```


````python
#include <stdio.h>

// Fonction pour calculer la facture d'électricité
// Cette fonction prend en entrée la consommation en kWh et applique les tarifs définis
double calculer_facture(double consommation_kwh) {
    double cout_total = 0.0;

    // Si la consommation est inférieure ou égale à 100 kWh, appliquer le tarif de 0,15 € par kWh
    if (consommation_kwh <= 100) {
        cout_total = consommation_kwh * 0.15;
    } 
    // Si la consommation est comprise entre 101 et 500 kWh, appliquer le tarif progressif
    else if (consommation_kwh <= 500) {
        cout_total = (100 * 0.15) + ((consommation_kwh - 100) * 0.20);
    } 
    // Si la consommation dépasse 500 kWh, appliquer un tarif pour chaque tranche
    else {
        cout_total = (100 * 0.15) + (400 * 0.20) + ((consommation_kwh - 500) * 0.25);
    }

    // Retourner le coût total calculé
    return cout_total;
}

int main() {
    // Variable pour stocker la consommation de l'utilisateur en kWh
    double consommation;

    // Affichage d'une invitation à l'utilisateur
    printf("Veuillez entrer votre consommation électrique en kWh : ");
    
    // Lecture de la consommation saisie par l'utilisateur
    scanf("%lf", &consommation);

    // Vérification : la consommation ne peut pas être négative
    if (consommation < 0) {
        printf("Erreur : la consommation ne peut pas être négative.\n");
        return 1; // Retourne un code d'erreur et termine le programme
    }

    // Appel de la fonction pour calculer le montant de la facture
    double facture = calculer_facture(consommation);

    // Affichage du résultat : la facture totale
    printf("Votre facture d'électricité est de : %.2f €\n", facture);

    return 0; // Fin du programme avec succès
}
````

### **Ajouts possibles**

1. **Gestion du stock** :

````python
#include <stdio.h>

// Fonction pour afficher le menu des boissons
void afficher_menu() {
    printf("=== Bienvenue à la machine à café ===\n");
    printf("1. Café Noir - 1.50 €\n");
    printf("2. Cappuccino - 2.50 €\n");
    printf("3. Thé - 1.20 €\n");
    printf("4. Chocolat Chaud - 2.00 €\n");
    printf("=====================================\n");
    printf("Choisissez une boisson en entrant le numéro correspondant :\n");
}

// Fonction pour retourner le prix d'une boisson choisie
float selectionner_boisson(int choix) {
    switch (choix) {
        case 1:
            return 1.50; // Café Noir
        case 2:
            return 2.50; // Cappuccino
        case 3:
            return 1.20; // Thé
        case 4:
            return 2.00; // Chocolat Chaud
        default:
            printf("Choix invalide. Veuillez réessayer.\n");
            return 0.0; // Aucun choix valide
    }
}

// Fonction pour gérer le paiement
void payer(float montant, float prix) {
    if (montant < prix) {
        printf("Montant insuffisant. Veuillez insérer %.2f € de plus.\n", prix - montant);
    } else {
        float monnaie = montant - prix;
        printf("Paiement accepté. Préparation de votre boisson...\n");
        if (monnaie > 0) {
            printf("Voici votre monnaie : %.2f €\n", monnaie);
        }
        printf("Merci pour votre achat !\n");
    }
}

int main() {
    int choix, continuer = 1;
    float montant, prix;

    while (continuer) {
        afficher_menu();
        scanf("%d", &choix); // Lecture du choix de l'utilisateur

        prix = selectionner_boisson(choix);
        if (prix == 0.0) {
            continue; // Si choix invalide, on redemande
        }

        printf("Le prix de la boisson choisie est : %.2f €\n", prix);
        printf("Veuillez insérer l'argent : ");
        scanf("%f", &montant);

        while (montant < prix) {
            printf("Montant insuffisant. Il manque %.2f €.\n", prix - montant);
            printf("Veuillez insérer de l'argent supplémentaire : ");
            float ajout;
            scanf("%f", &ajout);
            montant += ajout;
        }

        payer(montant, prix);

        printf("Souhaitez-vous acheter une autre boisson ? (1 : Oui, 0 : Non) : ");
        scanf("%d", &continuer);
    }

    printf("Merci d'avoir utilisé notre machine à café. À bientôt !\n");
    return 0;
}
````
