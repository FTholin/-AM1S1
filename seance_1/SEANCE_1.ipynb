## **Comment fonctionne un ordinateur ?**


Interagir avec un ordinateur est une expérience très interactive :

- **Lorsque nous tapons sur un clavier**, l'ordinateur reconnaît et affiche les caractères correspondants.
- **Lorsque nous bougeons la souris**, l'ordinateur crée un curseur qui suit ces mouvements.

Mais comment un ordinateur peut-il comprendre et répondre à nos actions ?

Dans cette leçon, nous explorerons les quatre fonctions principales qui permettent à un ordinateur de répondre à nos interactions :

- **Entrée** : Comment les actions comme les clics et les frappes sont captées.
- **Traitement** : L'analyse et la réponse aux entrées.
- **Mémoire** : Où les informations sont stockées temporairement ou de manière permanente.
- **Sortie** : Comment les résultats sont affichés ou reproduits.

Chaque fonction est cruciale pour permettre à l'ordinateur de fonctionner de manière efficace et réactive.

### Comprendre les entrées d'un ordinateur

Les entrées sur un ordinateur proviennent de nos interactions quotidiennes, comme cliquer sur une souris ou ajuster le volume. Ce sont des commandes physiques transformées en données numériques que l'ordinateur peut comprendre et traiter.

Voici comment cela fonctionne :

1. **Connexion** : Les dispositifs d'entrée (souris, clavier, etc.) sont connectés à l'ordinateur, généralement via un port sur la carte mère.
2. **Détection et signalisation** : Le dispositif d'entrée détecte nos actions (comme un clic) et envoie un signal au processeur central (CPU).
3. **Conversion en binaire** : Le signal est converti en code binaire, un langage composé uniquement de 0 et de 1, que le CPU peut traiter.

En bref, nos actions physiques sont converties en instructions numériques que l'ordinateur exécute.


### Traitement des données



Après avoir reçu des données, notre ordinateur doit les traiter pour comprendre nos demandes et savoir comment y répondre. Cette mission est dévolue au processeur central, ou CPU, le "cerveau" de l'ordinateur.

#### Rôle du CPU
Le CPU joue un rôle central, gérant la communication entre le matériel (hardware) et le logiciel (software) :
- **Interprétation des actions** : Si nous augmentons le volume sur nos haut-parleurs, le CPU reçoit et interprète ces données.
- **Exécution des logiciels** : Lorsque nous lançons un logiciel, le CPU traite toutes les opérations nécessaires pour que le logiciel fonctionne correctement.

Avec la multitude de logiciels et de processus que nous utilisons, le CPU doit traiter les informations rapidement. C’est là que la mémoire de l'ordinateur intervient pour aider à la gestion efficace de ces données.

### Gestion de la mémoire dans les ordinateurs
La mémoire d'un ordinateur est essentielle pour stocker et accéder aux informations, que ce soit temporairement ou de manière permanente. Comment cela fonctionne-t-il ?

#### Mémoire Temporaire (RAM)
La mémoire temporaire, ou RAM, est utilisée pour des informations dont nous avons besoin brièvement. Par exemple, le numéro de téléphone d'un restaurant ne doit être retenu que le temps de passer une commande. De même, lorsqu'un logiciel est en cours d'utilisation, la RAM stocke temporairement les instructions nécessaires pour que le CPU puisse les exécuter rapidement. Cependant, une fois le programme fermé ou l'ordinateur éteint, ces données disparaissent.

#### Mémoire Permanente (Disque dur)
Pour des données à conserver sur le long terme, comme des centaines de photos, nous utilisons la mémoire permanente. Ces données sont généralement sauvegardées sur le disque dur de l'ordinateur, permettant un accès durable et répété à ces informations.

En résumé, la mémoire d'un ordinateur travaille de concert pour gérer efficacement les informations nécessaires à court terme via la RAM, et celles à conserver durablement sur le disque dur.

### Sortie


Une fois que le CPU a traité les données et envoyé des instructions sur leur gestion, une sortie est produite ! La sortie peut être vue comme l'exécution de la commande que nous avons donnée via un dispositif de sortie.

Voici quelques exemples de dispositifs de sortie :

- **Moniteurs**
- **Haut-parleurs**
- **Imprimantes**

Par exemple, si nous cliquons sur une souris pour ouvrir un fichier, la sortie sera le contenu du fichier affiché sur le moniteur. Si nous augmentons le volume de nos haut-parleurs, la sortie sera le son qui devient plus fort.

La sortie représente l'étape finale dans le processus d'interaction avec notre ordinateur.

## Système de Numération Binaire

### Introduction
Comme le dit la vieille blague :

***Il y a seulement 10 types de personnes dans le monde, celles qui comprennent le binaire et celles qui ne le comprennent pas.***

Le binaire est défini comme quelque chose qui concerne, est composé de, ou implique deux éléments.

En informatique, les exemples de binaire se divisent en deux catégories distinctes :

- **Nombres Binaires**
- **Données Binaires**

Les nombres binaires sont exprimés par une combinaison de 0 et de 1. Par exemple, 100110 est l'équivalent binaire du nombre 38.

Parmi les exemples courants de données binaires, on trouve :

- **Code Machine** (001010101100111001010010011)
- **Expressions Booléennes** (Vrai ou Faux)
- **États Matériels** (Allumé ou Éteint)
- **Réseautique et Stockage de Fichiers**

Avant de plonger dans l'utilisation des données, prenons le temps de vraiment comprendre ce que sont les données binaires et le système de numération binaire.

### Ordinateurs et Système Binaire

Les ordinateurs fonctionnent avec le système binaire, utilisant uniquement deux états, off (0) et on (1), car gérer dix états différents comme dans le système décimal serait trop coûteux et complexe. Cette simplicité permet aux ordinateurs de traiter les données efficacement.

Les données binaires sont traitées soit comme de l'énergie appliquée, soit comme de l'énergie non appliquée, nécessitant une grande précision dans la gestion de la tension électrique. Les données ont des longueurs spécifiques, comme les octets (8 bits) et les mots (16 bits), facilitant leur traitement par le matériel.

Par exemple, la vitesse d'internet est mesurée en bits par seconde, tandis que la taille des fichiers est en octets. Un gigabit de vitesse internet (1 Gb/s) paraît rapide, mais comme un octet équivaut à huit bits, le téléchargement prendra en réalité huit fois plus de temps que prévu.

En résumé, tout, du texte à l'image sur votre écran, est représenté par des combinaisons de 0 et de 1.

### Système de Numération Binaire

Les nombres ont été représentés de différentes manières à travers l'histoire. Par exemple, sur certaines horloges, vous pouvez voir que six heures est indiqué par VI, le chiffre romain pour 6.

Le système de numération le plus réussi est appelé le système décimal, du mot latin "dec-" signifiant ensemble de dix ou ayant une base de dix.

Bien que les origines exactes de ce système soient inconnues, il est clair qu'il a commencé avec le comptage sur nos doigts et a plus tard évolué en substituant les caractères hindou-arabes de 0, 1, 2, 3, 4, 5, 6, 7, 8 et 9 aux doigts pour effectuer des opérations plus importantes.

Dans le système décimal, chaque chiffre peut être représenté par un multiple d'une puissance de dix et ajouté aux autres chiffres. Prenons le nombre 305.

## 1. Puissances et Règles de Calcul

**Définition et exemples :**

- **Puissance positive :** Multiplier un nombre par lui-même plusieurs fois.
  - Exemples :
    - $2^3 = 2 \times 2 \times 2 = 8$
    - $4^2 = 4 \times 4 = 16$

- **Puissance zéro :** Tout nombre élevé à la puissance zéro est égal à 1.
  - Exemple : $5^0 = 1$

- **Puissance négative :** Correspond à l'inverse de la puissance positive.
  - Exemple : $3^{-2} = \frac{1}{3^2} = \frac{1}{9}$

**Règles de calcul des puissances :**

- **Addition des exposants lors de la multiplication :**
  - $x^{m+n} = x^m \times x^n$
  - Exemple : $2^3 \times 2^4 = 2^{3+4} = 2^7 = 128$

- **Soustraction des exposants lors de la division :**
  - $\frac{x^m}{x^n} = x^{m-n}$
  - Exemple : $\frac{5^4}{5^2} = 5^{4-2} = 5^2 = 25$

- **Multiplication des exposants :**
  - $(x^m)^n = x^{m \times n}$
  - Exemple : $(2^3)^2 = 2^{3 \times 2} = 2^6 = 64$

## 2. Division Euclidienne

**Concept :** La division euclidienne divise un nombre $a$ par un diviseur $b$, donnant un quotient $q$ et un reste $r$ tel que 

$a = bq + r$ 

avec 

$0 \leq r < b$.

**Exemples :**

- Diviser 20 par 3 :
  - $20 = 3 \times 6 + 2$
  - Quotient : 6, Reste : 2

- Diviser 50 par 7 :
  - $50 = 7 \times 7 + 1$
  - Quotient : 7, Reste : 1

**Applications :** 
- Calculs manuels pour vérifier des divisions.
- Utilisation en programmation (Python) pour trouver le quotient et le reste :
  - Quotient : `q = a // b`
  - Reste : `r = a % b`
  - Exemple en Python : `20 // 3` retourne 6 et `20 % 3` retourne 2.

## 3. Numération et Bases

**Bases courantes :**

- **Base 2 (Binaire) :** Utilise les chiffres 0 et 1, essentiel en informatique.
  - Exemple : $(1010)_2 = 1 \times 2^3 + 0 \times 2^2 + 1 \times 2^1 + 0 \times 2^0 = (10)_{10}$

- **Base 10 (Décimale) :** Système standard utilisé dans la vie quotidienne.

- **Base 16 (Hexadécimale) :** Utilise les chiffres 0-9 et les lettres A-F pour représenter 10 à 15.
  - Exemple : $(1A)_{16} = 1 \times 16^1 + 10 \times 16^0 = 26$

Ces bases sont essentielles pour comprendre les systèmes numériques, notamment en programmation et en électronique.

## 4. Conversion entre les Bases

**Conversion d’une base vers le décimal :**

- Multipliez chaque chiffre par sa position pondérée par la base.
- Exemple : Convertir $(1011)_2$ en décimal :

  $(1011)_2 = 1 \times 2^3 + 0 \times 2^2 + 1 \times 2^1 + 1 \times 2^0 = 8 + 0 + 2 + 1 = (11)_{10}$

**Conversion du décimal vers une autre base (Méthode des divisions successives) :**

1. Divisez le nombre par la base cible.
2. Notez le reste.
3. Continuez à diviser le quotient jusqu’à obtenir 0.
4. Lisez les restes de bas en haut pour obtenir le nombre dans la nouvelle base.

- **Exemple :** Convertir $(45)_{10}$ en base 2 :
  - 45 ÷ 2 = 22, reste 1
  - 22 ÷ 2 = 11, reste 0
  - 11 ÷ 2 = 5, reste 1
  - 5 ÷ 2 = 2, reste 1
  - 2 ÷ 2 = 1, reste 0
  - 1 ÷ 2 = 0, reste 1
  - Résultat : $(101101)_2$

Ces conversions sont cruciales pour la compréhension et l’utilisation des différentes bases en programmation, électronique, et calculs numériques.

**Exercices :**
1. Convertissez $(1010)_2$ en base 10.
2. Convertissez $(37)_{10}$ en base 8.
3. Transformez $(255)_{10}$ en base 16.

<details>
  <summary><strong>Voir les réponses</strong></summary>

  **Réponses :**
  1. $(1010)_2 = 1 \times 2^3 + 0 \times 2^2 + 1 \times 2^1 + 0 \times 2^0 = 10$.
  2. $37 \div 8 = 4$ reste $5$, $4 \div 8 = 0$ reste $4$. Donc, $(37)_{10} = (45)_8$.
  3. $255 \div 16 = 15$ reste $15$, donc $(255)_{10} = (FF)_{16}$.

</details>

## 5. Opérations en Binaire

**Addition binaire :** Similaire à l’addition décimale, avec des retenues à 1.

- **Exemple :** $(1010)_2 + (1101)_2$
  - Alignement : 
    ```
      1010
    + 1101
    ------
     10111
    ```
  - Résultat : $(10111)_2$

**Multiplication binaire :** Semblable à la multiplication en base 10 mais avec 0 et 1.

- **Exemple :** $(101)_2 \times (11)_2$
  - Alignement :
    ```
      101
    ×  11
    -----
      101
    +1010
    -----
     1111
    ```
  - Résultat : $(1111)_2$
  
  
**Exercices :**
1. Additionnez les nombres binaires $(1011)_2$ et $(1101)_2$.
2. Multipliez $(110)_2$ par $(11)_2$.
3. Additionnez $(111)_2$ et $(101)_2$.

<details>
  <summary><strong>Voir les réponses</strong></summary>

  **Réponses :**
  1. $(1011)_2 + (1101)_2 = (11000)_2$.
  2. $(110)_2 \times (11)_2 = (10010)_2$.
  3. $(111)_2 + $(101)_2 = (1100)_2$.

</details>


Ces opérations sont au cœur de la logique numérique, des algorithmes et des circuits digitaux.
